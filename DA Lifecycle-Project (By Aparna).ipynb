## Data Collection

```python
import pandas as pd  #for data handling and analysis
```

```python
df = pd.read_csv("train.csv") # Load dataset from csv file into dataframe
df #displays the dataset
```

```python
df.head(10)  # Display first 10 rows of the dataset
```

```python
df.tail(10) #Display last 10 rows of the dataset
```

### Data Cleaning

```python
df.isnull().sum() # Checks and counts missing (null) values in each column
```

```python
df.columns  # Displays all column names in the dataset
```

```python
df=df.drop(["Alley","PoolQC","Fence","MiscFeature"],axis=1,errors='ignore')
# drop(): removes columns
# axis=1: column-wise operation
# errors='ignore': avoids error if column is not present
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].median(),inplace=True)
# fillna(): fills missing values
# median(): middle value (used for numerical data)
# Fills missing values in LotFrontage column using the median value

df['GarageYrBlt'].fillna(0,inplace=True)  #0: used when value is not available
# Fills missing values in GarageYrBlt column with 0
```

## Data Preprocessing

```python
df=pd.get_dummies(df, drop_first=True) # Convert categorical to numerical
# get_dummies(): converts text data into numbers
# drop_first=True: removes one category to avoid duplicate columns
```

```python
df.dtypes #check data types
```

```python
X=df.drop('SalePrice',axis=1) # Creates feature variables by removing the target column (SalePrice)
X
```

```python
y=df['SalePrice'] # Stores the target variable (SalePrice) separately 
y
```

## EDA (Exploratory Data Analysis)

### Univariate Analysis

```python
df['SalePrice'].describe() #Non visualization
#describe(): gives summary statistics
```

```python
import matplotlib.pyplot as plt # for plotting graphs

plt.hist(df['SalePrice']) # Creates a histogram to show distribution of SalePrice
plt.title("Distribution of Sale Price") # Adds title to the graph
plt.show()  # Displays the graph             #Visualization
```

### Bivariate Analysis

```python
df[['SalePrice','GrLivArea']].corr()  #Non visualization
# Calculates correlation between SalePrice and GrLivArea
#corr(): shows relationship between variables

#+1: strong positive relationship (increase to increase)
#0: no relationship
#-1: strong negative relationship (increase to decrease)
```

```python
plt.scatter(df['GrLivArea'],df['SalePrice']) # Creates a scatter plot between Living Area and Sale Price
plt.xlabel("Living Area") # Labels X-axis (independent variable)
plt.ylabel("Sale Price") # Labels Y-axis (dependent variable)
plt.title("Area vs Price")  # Adds title to the graph
plt.show()   # Displays the graph     #Visualization
```

### Multivariate Analysis

```python
df.corr() #Non visualization
# Calculates correlation between all numerical columns in the dataset
```

```python
import seaborn as sns #for advanced visualizations

corr = df.corr() # Calculates correlation between all numerical columns

top_corr = corr['SalePrice'].abs().sort_values(ascending=False).head(15)
# Selects top 15 features most correlated with SalePrice
# abs(): Converts values to positive
# sort_value: Sorts correlation from highest to lowest

corr_matrix = df[top_corr.index].corr() #Creates a smaller correlation matrix using selected columns

plt.figure(figsize=(12,10)) # Sets the size of the figure
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm') #Creates heatmap with correlation values displayed
plt.title("Correlation Heatmap") # Adds title to the graph
plt.show() # Displays the heatmap   #Visualization
```

## Observations/ Insights
